```python
import numpy as np
numbers = np.array([1,3,5,7,9])
print("Array:",numbers)

result = numbers + 2
print("new array:", result)

result_2 = result * 3
print("updated array:", result_2)

print("sum of original array :", np.sum(numbers))
print("mean of the original array:", np.mean(numbers))
print("maximum of the original array:", np.max(numbers))
```
