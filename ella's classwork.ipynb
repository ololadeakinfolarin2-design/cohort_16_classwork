## creating a NumPy

```python
import numpy as np
```

```python
nums= np.array([1,3,5,7,9])
print("array:",nums)
```

```python
nums_2=nums + 2
print("add_array:",nums_2)
```

```python
nums_3=nums * 3
print("multiplied_array:",nums_3)
```

```python
nums =np.sum(nums)
print(nums)
```

```python
nums=np.max(nums)
print(nums)
```

```python
nums=np.mean(nums)
print(nums)
```
